### __Statement 1__:

To explore how the Fama-French Five-Factor Model explains the daily returns of technology companies througout the entire testing period (2021-2024), we began by testing which factors in the model are most influential. This analysis allows us to identify whether key elements like market excess return (Mkt-RF) or size (SMB) have a measurable impact on stock performance. Based on this, we formulated the following hypotheses:


- $H_0$: Each of the Fama-French factors,Mkt-RF, SMB, HML,RMW, and CMA have no impact on the daily returns of the selected technolohgy companies. In terms of regression coefficients: $\beta_{\text{Mkt-RF}} = \beta_{\text{SMB}} = \beta_{\text{HML}} = \beta_{\text{RMW}} = \beta_{\text{CMA}} = 0$


- $H_A$: Each of the Fama-French factors,Mkt-RF, SMB, HML,RMW, and CMA  have a significant on the daily returns of the chosen technology companies. In terms of the regression coefficient: $\beta_{\text{Mkt-RF}} \neq 0$ , $\beta_{\text{SMB}} \neq 0$, $\beta_{\text{HML}} \neq 0$, $\beta_{\text{RMW}} \neq 0$,  $\beta_{\text{CMW}} \neq 0$


We will evaluate the null hypothesis by fitting a regression model using the Fama-French factors as independent variables and the daily returns of technology companies as the dependent variable. For each factor, we will examine the regression coefficient $\beta$ and its corresponding p-value. A p-value less than the significance threshold of p < 0.05 indicates that the factor has a significant impact on returns, leading us to reject the null hypothesis for that factor. This approach ensures we identify which factors are statistically influential during the AI boom.


### Data Analysis for Statement 1:

```python
# Rename Mkt-RF for ease of use 
merged_df = merged_df.rename(columns={'Mkt-RF': 'Mktrf'})

y = merged_df['Daily_Return']  # Dependent variable (e.g., daily stock returns)
X = merged_df[['Mktrf', 'SMB', 'HML', 'RMW', 'CMA']]  # Independent variables

X = sm.add_constant(X)

# Fit the OLS Model 
model = sm.OLS(y, X).fit()
# Print the Model Summary
print(model.summary())
```

### Printing the p-values more clearly:

```python

for factor in ['Mktrf', 'SMB', 'HML', 'RMW', 'CMA']:
    beta = model.params[factor]
    p_value = model.pvalues[factor]
    print(f"Beta ({factor}): {round(beta,6)}")
    print(f"P-Value ({factor}): {p_value}")
```

### Evaluation of Significance:


$\text{Mkt-RF}$: The regression analysis yielded $\beta_{\text{Mkt-RF}} = 0.012$, indicating a positive relationship between the market risk premium and the daily returns of the selected technology companies. The associated p-value is $p = 0.0$, which is far below the significance threshold of $p < 0.05$. Therefore we can reject the null hypothesis that $\beta_{\text{SMB}} \neq 0 $




$\text{SMB}$: $\beta_{\text{SMB}} = -0.0018$, indicating a negative relationship between the size factor and the daily returns. The associated p-value is $p = 4.13 \times 10^{-5}$, which is below the significance threshold of $p < 0.05$. Therefore we can reject the null hypothesis that $\beta_{\text{SMB}} \neq 0 $


$\text{HML}$:  $\beta_{\text{HML}} = -0.0029$, indicating a negative relationship between the value factor and the daily returns. The associated p-value is $p = 5.32 \times 10^{-13}$, which is below the significance threshold of $p < 0.05$. Therefore we can reject the null hypothesis that $\beta_{\text{SMB}} \neq 0 $


$\text{RMW}$: The analysis showed that $\beta_{\text{RMW}} = -0.0012$, indicating a negative relationship between the factor and the daily returns. The associated p-value is $p = 0.014$, which is below the significance threshold of $p < 0.05$. Therefore we can reject the null hypothesis that $\beta_{\text{RMW}} \neq 0 $


$\text{CMA}$: The $\beta_{\text{CMA}} = -0.0041$, indicating a negative relationship between the investment factor and the daily returns. The associated p-value is $p = 1.36 \times 10^{-10}$, which is far below the significance threshold of $p < 0.05$. Therefore we can reject the null hypothesis that $\beta_{\text{RMW}} \neq 0 $

In summary, the above analysis shows that all of the Fama-French Five factors, $\text{Mkt-RF}$, $\text{SMB}$, $\text{HML}$, $\text{RMW}$, and $\text{CMA}$ have a statistically significant impact on the daily returns of the technology companies. This regression also highlights that the Market Risk Premium ($\text{Mkt-RF}$) is the most significant factor influencing daily returns for the chosen technology companies as it has a positive beta, and the smallest calculated p-value.

### __Statement 2:__

A key question in understanding tech stocks during the AI boom is whether they became more sensitive to market-wide trends, as captured by the market excess return (Mkt-Rf). To explore this, we are comparing how tech stocks responded to Mkt-Rf during two one-year periods: 2021-22 (pre-AI boom) and 2023-24 (AI boom). Using consistent time frames allows us to isolate changes caused by market conditions rather than seasonal fluctuations. We chose to focus on Mkt-Rf because our previous analysis showed it had the strongest correlation with daily returns among the Fama-French Five Factors, making it the most relevant factor for this analysis.

To visualize this, we’ll create a graph where the x-axis represents time intervals (e.g., months), and the y-axis reflects sensitivity to Mkt-Rf. Lines representing the two periods will be color-coded for clarity. We will use hypothesis testing, including a Z-test and p-value analysis, to determine whether the differences in sensitivity between the two periods are statistically significant. This analysis will help us understand whether the AI boom amplified tech stocks’ responsiveness to market-wide movements and how speculative trends can influence stock behavior. 

#### Hypotheses: 

- Null Hypothesis ($H_0$): The sensitivity of tech stocks to market-wide movements (as measured by the correlation between $\text{Mkt-RF}$ and daily returns) is the <b> same </b> during the pre-AI boom period (2021-2022) and the AI boom period (2023-2024). 


- Alternative Hypothesis ($H_A$): The sensitivity of tech stocks to market-wide movements is <b> not the same </b> between the pre-AI boom period (2021-2022) and the AI boom period (2023-2024).  


### Data Analysis for Statement 2:

```python


# Step 1: Split the merged_df into two time periods
period1 = merged_df[(merged_df['Date'] >= '2021-01-01')
& (merged_df['Date'] <= '2022-12-31')].copy()
period2 = merged_df[(merged_df['Date'] >= '2023-01-01') 
& (merged_df['Date'] <= '2024-12-07')].copy()


period1['Date'] = pd.to_datetime(period1['Date'])
period2['Date'] = pd.to_datetime(period2['Date'])


# Find correlations
correlation1 = period1['Mktrf'].corr(period1['Daily_Return'])
correlation2 = period2['Mktrf'].corr(period2['Daily_Return'])


print("Correlation 1:", correlation1)
print("Correlation 2:", correlation2)


period1 = period1.sort_values(by='Date')
period2 = period2.sort_values(by='Date')


period1_filtered = period1.set_index('Date').\
resample('3ME').first().reset_index()
period2_filtered = period2.set_index('Date').\
resample('3ME').first().reset_index()


# Plotting
plt.figure(figsize=(12, 6))
plt.plot(period1['Date'], period1['Mktrf'], label='Before AI',
        color='darkseagreen', linestyle='--', marker='o')
plt.plot(period2['Date'], period2['Mktrf'], label='After AI',
        color='palevioletred', linestyle='-', marker='o')


plt.xlabel('Time (Months)')
plt.ylabel('Mkt-Rf')
plt.title('Mkt-Rf Before AI vs. After AI')
plt.legend()
plt.show()


# Hypothesis Testing
n1 = len(period1)
n2 = len(period2)


z1 = 0.5 * np.log((1 + correlation1) / (1 - correlation1))
z2 = 0.5 * np.log((1 + correlation2) / (1 - correlation2))


se1 = 1 / np.sqrt(n1 - 3)
se2 = 1 / np.sqrt(n2 - 3)


z_stat = (z1 - z2) / np.sqrt(se1**2 + se2**2)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))


# Results
print(f"Fisher Z-Value Period 1: {z1}," 
f"Fisher Z-Value Period 2: {z2}")
print(f"P-Value: {p_value}")
```

### Evaluation of Significance: 

To evaluate the impact of the AI boom on the sensitivity of tech stocks to market-wide movements, we began by dividing the data into two time periods: January 1, 2021, to December 1, 2022, representing the pre-AI boom, and January 1, 2023, to December 1, 2024, representing the AI boom. This consistent one-year division allowed us to isolate changes driven by market conditions rather than seasonal fluctuations, focusing specifically on the speculative growth period of AI. Correlation coefficients were then calculated for each period to measure the relationship between market excess return (Mkt-Rf) and daily returns. For the pre-AI period, the correlation was 0.641, indicating a relatively strong relationship between market-wide trends and tech stock returns. During the AI boom, the correlation dropped to 0.510, suggesting that tech stocks became less sensitive to broader market movements.

To visually compare these periods, we plotted Mkt-Rf over time, with the x-axis representing monthly intervals and the y-axis showing Mkt-Rf. Distinct line styles and colors were used to differentiate the two periods: a red dashed line for the pre-AI boom and a green solid line for the AI boom. This visualization provided a clear comparison of market-wide trends across the two periods and highlighted any notable differences in behavior or volatility. Sensitivity was further analyzed using Fisher Z-transformation, which standardized the correlation coefficients and enabled statistical comparison of the two periods.

Hypothesis testing was conducted to determine if the sensitivity of tech stocks to Mkt-Rf changed significantly between the two periods. The <b> null hypothesis </b> assumed that sensitivity was equal across both periods while the <b> alternative hypothesis </b> proposed that sensitivity was not equal. Using the Fisher Z-transformed values for each period—0.760 for 2021-2022 and 0.563 for 2023-2024—a Z-statistic and p-value were calculated. The resulting p-value was 3.11×10−153.11 \times 10^{-15}3.11×10−15, which is far below the 0.05 significance threshold, leading to the rejection of the null hypothesis. This indicates a statistically significant change in sensitivity, with tech stocks becoming less responsive to market-wide movements during the AI boom.

These findings suggest that the AI boom introduced unique factors influencing tech stock performance, reducing their dependence on overall market trends. This shift may be attributed to heightened speculative interest, company-specific developments, or other dynamics tied to AI adoption. The results underscore how major technological trends can reshape traditional market dynamics, providing critical insights for investors and analysts alike. By combining correlation analysis, visualization, and rigorous hypothesis testing, this evaluation highlights the significant impact of the AI boom on the behavior of tech stocks.


###  __Statement 3:__

To investigate whether the market risk premium (Mkt-RF) influences AI-focused companies differently than general technology companies, we include an interaction term between Mkt-RF and AI-Focus in our model. This allows us to test whether AI-focused companies, such as Nvidia and Palantir, respond to market-wide trends differently compared to general tech companies like Microsoft and IBM. Based on this, we define the following hypotheses:


- Null Hypothesis ($H_0$):
The relationship between $\text{Mkt-RF}$ and daily returns is the same for AI-focused companies (Nvidia, Palantir) and general technology companies (Microsoft, IBM): $\beta_{\text{3}} = 0$
  
   Where $\beta_{\text{3}}$  is the coefficient of the interaction term $\text{Mkt-RF}$ $\times$ $\text{AI-Focus}$.


- Alternative Hypothesis ($H_A$):
The relationship between $\text{Mkt-RF}$ and daily returns differs for AI-focused companies and general technology companies: $\beta_{\text{3}} \neq 0$
A significant $\beta_{\text{3}}$ would indicate that AI-focused companies respond differently to market risk premium changes compared to general technology companies.




To test the null hypothesis , we will assess the p-value for the interaction term. If the p-value is below the significance threshold ( p < 0.05), we will reject the null hypothesis, concluding that the relationship between Mkt-RF and returns differs for AI-focused companies. Otherwise, we will fail to reject the null hypothesis, suggesting no significant difference.





### Data Analysis for Statement 3:

```python
# Add an AI_Focus column (1 for AI-specific, 0 for general tech)
ai_companies = ['NVDA', 'PLTR']  # AI-specific companies
merged_ai_df = merged_df.copy()
merged_ai_df['AI_Focus'] = merged_ai_df['Ticker'].apply(\
    lambda x: 1 if x in ai_companies else 0)

import statsmodels.formula.api as smf

# Fit the regression model with an interaction term
model_2 = smf.ols(
    'Daily_Return ~ Mktrf + AI_Focus + \
    Mktrf:AI_Focus + SMB + HML + RMW + CMA',
    data=merged_ai_df
).fit()

# Print the regression summary
print(model_2.summary())
```

### Evaluation of Significance:


The interaction analysis reveals that the relationship between market risk premium ($\text{Mkt-RF}$) and daily returns differs significantly between AI-focused companies (Nvidia, Palantir) and general technology companies (e.g., IBM, Microsoft). The positive and statistically significant interaction term ($\beta_3 = 0.0117$, $p < 0.00$) reveals that AI-focused companies demonstrate a 1.17% stronger impact from ($\text{Mkt-RF}$) on their daily returns compared to the general tech firms after controlling for the other factors ($\text{SMB}$, $\text{HML}$, $\text{RMW}$, and $\text{CMA}$). This finding supports the hypothesis that AI-focused companies possibly are more sensitive to broader market movements as outlined by the Fama-French factors, likely due to perceived growth potential during the AI boom.

```python
ai_companies = ['NVDA', 'PLTR']
general_tech_companies = ['IBM', 'MSFT']
selected_companies = ai_companies + general_tech_companies

filtered_df = merged_df[merged_df['Ticker'].\
isin(selected_companies)].copy()


# Add AI_Focus column
filtered_df['AI_Focus'] = filtered_df['Ticker'].apply(\
    lambda x: 1 if x in ai_companies else 0)

#Run model and print 
model_2 = smf.ols(
   'Daily_Return ~ Mktrf + AI_Focus + \
   Mktrf:AI_Focus + SMB + HML + RMW + CMA',
   data=filtered_df
).fit()


print(model_2.summary())
```

- The insignificance of the RMW (Robust Minus Weak) factor in our analysis is likely due to the dataset being composed entirely of large, well-established tech companies with uniformly high and stable profitability. This lack of variability in profitability diminishes the factor's ability to explain differences in returns, even though it may play a more significant role in a broader or more diverse portfolio.

- Although the SMB (Small Minus Big) factor was statistically significant, its findings should be interpreted with caution because all companies in our dataset are large-cap tech firms. The lack of smaller firms in the sample skews the SMB factor, as there is no true variation in size to reflect its impact. As a result, SMB’s significance may not provide meaningful insights in this specific analysis.

- One big limitation of our study is the impact of the COVID-19 pandemic, which happened just a year before our dataset starts. The pandemic caused a major recession and had a huge effect on the economy, with recovery lasting into 2021 and 2022. Since this overlaps with the pre-AI boom period in our analysis, it’s possible that the market dynamics during this recovery period influenced our results. This makes it harder to tell how much of the changes we see are due to the AI boom versus the lingering effects of the pandemic.
